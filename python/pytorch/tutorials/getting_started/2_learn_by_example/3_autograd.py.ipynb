```python
import torch
```

```python
dtype = torch.float
device = torch.device('cpu')
```

```python
BATCH, N_IN, HIDDEN, N_OUT = 64, 1000, 100, 10
ETA = 1e-6
```

```python
x = torch.randn(BATCH, N_IN, device=device, dtype=dtype)
y = torch.randn(BATCH, N_OUT, device=device, dtype=dtype)
```

```python
w1 = torch.randn(
    N_IN, HIDDEN, device=device, dtype=dtype, requires_grad=True)
w2 = torch.randn(
    HIDDEN, N_OUT, device=device, dtype=dtype, requires_grad=True)
```

```python
for t in range(500):
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    loss = (y_pred - y).pow(2).sum()
    print(t, loss.item())
    loss.backward()
    with torch.no_grad():
        w1 -= ETA * w1.grad
        w2 -= ETA * w2.grad
        
        # Zero-out gradients after updating weights
        w1.grad.zero_()
        w2.grad.zero_()
```
