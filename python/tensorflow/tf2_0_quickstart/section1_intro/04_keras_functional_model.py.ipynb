```python
import tensorflow as tf
from   tensorflow.keras import Input, Model
from   tensorflow.keras.layers import Dense, Dropout, Flatten
from   tensorflow.keras.optimizers import Adam
```

```python
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train / 255.
X_test = X_test / 255.
```

```python
EPOCHS = 10
DROPOUT = 0.2
BATCH = 32
```

```python
inputs = Input(shape=(28, 28)) # returns placeholder tensor
X = Flatten()(inputs)
X = Dense(512, activation='relu', name='d1')(X)
X = Dropout(DROPOUT)(X)
preds = Dense(10, activation=tf.nn.softmax, name='d2')(X)
mod3 = Model(inputs=inputs, outputs=preds)
```

```python
mod3.summary()
```

```python
optim = Adam()
mod3.compile(optimizer=optim, 
             loss='sparse_categorical_crossentropy', 
             metrics=['accuracy'])
mod3.fit(X_train, y_train, batch_size=BATCH, epochs=EPOCHS)
mod3.evaluate(X_test, y_test)
```
