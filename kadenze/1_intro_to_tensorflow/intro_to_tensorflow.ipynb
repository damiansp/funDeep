```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

%matplotlib inline
fig_size = [16, 9]
plt.rcParams['figure.figsize'] = fig_size
```

```python
x = np.linspace(-3., 3., 100)
print(x.dtype)
print(x)
```

```python
x = tf.linspace(-3., 3., 100)
print(x)
```

```python
g = tf.get_default_graph()
[op.name for op in g.get_operations()]
```

```python
g.get_tensor_by_name('LinSpace' + ':0')
```

```python
s = tf.Session()
```

```python
x_computed = s.run(x)
print(x_computed)
```

```python
# Alternately
x_computed = x.eval(session = s)
print(x_computed)
```

```python
# Create some standard normal data
mean = 0.
sd = 1.

z = (tf.exp(tf.negative(tf.pow(x - mean, 2.) / (2. * tf.pow(sd, 2.)))) * 
     (1. / (sd * tf.sqrt(2. * np.pi))))
rnorm = z.eval()
plt.plot(rnorm);
```

```python
kernsize = z.get_shape().as_list()[0]
z2d = tf.matmul(tf.reshape(z, [kernsize, 1]), tf.reshape(z, [1, kernsize]))
plt.imshow(z2d.eval());
```
