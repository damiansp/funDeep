```python
data = np.array(imgs)
data.shape
```

```python
mean_img = np.mean(data, axis = 0)
median_img = np.median(data, axis = 0)

plt.imshow(mean_img.astype(np.uint8));
```

```python
plt.imshow(median_img.astype(np.uint8));
```

```python
mean_img = np.mean(data[:20], axis = 0)
median_img = np.median(data[:20], axis = 0)
plt.imshow(mean_img.astype(np.uint8));
```

```python
plt.imshow(median_img.astype(np.uint8));
```

```python
mean_img = np.mean(data[40:60], axis = 0)
median_img = np.median(data[40:60], axis = 0)
plt.imshow(mean_img.astype(np.uint8));
```

```python
plt.imshow(median_img.astype(np.uint8));
```

```python
std_img = np.std(data, axis = 0)
plt.imshow(std_img.astype(np.uint8));
```

```python
plt.imshow(np.mean(std_img, axis = 2).astype(np.uint8), cmap = 'gray');
```

```python
flattened = data.ravel()
```

```python
fig, axs = plt.subplots(
    1, 3, figsize = (12, 6), sharey = True, sharex = True)
axs[0].hist((data[0]).ravel(), 255);
axs[1].hist((mean_img).ravel(), 255);
axs[2].hist((data[0].ravel() - mean_img.ravel()), 255);
```
