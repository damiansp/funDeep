```python
import torch
```

```python
dtype = torch.float
device = torch.device('cpu')
# device = torc.device('cuda:0')
```

```python
BATCH, D_IN, HIDDEN, D_OUT = 64, 1000, 100, 10
```

```python
x = torch.randn(BATCH, D_IN, device=device, dtype=dtype)
y = torch.randn(BATCH, D_OUT, device=device, dtype=dtype)
```

```python
ETA = 1e-6
```

```python
w1 = torch.randn(D_IN, HIDDEN, device=device, dtype=dtype)
w2 = torch.randn(HIDDEN, D_OUT, device=device, dtype=dtype)
```

```python
for t in range(500):
    # forward
    h = x.mm(w1) # mm: matrix multiply
    h_relu = h.clamp(min=0)
    y_pred = h_relu.mm(w2)
    
    loss = (y_pred - y).pow(2).sum().item()
    print('%3d: %10.4f' % (t, loss))
    
    # Back
    grad_y_pred = 2. * (y_pred - y)
    grad_w2 = h_relu.t().mm(grad_y_pred)
    grad_h_relu = grad_y_pred.mm(w2.t())
    grad_h = grad_h_relu.clone()
    grad_h[h < 0] = 0
    grad_w1 = x.t().mm(grad_h)
    
    # Update Ws
    w1 -= ETA * grad_w1
    w2 -= ETA * grad_w2
```
