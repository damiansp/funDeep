```python
import numpy as np
```

```python
BATCH = 64
D_IN = 1000
HIDDEN = 100
D_OUT = 10
```

```python
# Random x, y data
x = np.random.randn(BATCH, D_IN)
y = np.random.randn(BATCH, D_OUT)
```

```python
# Init weights
w1 = np.random.randn(D_IN, HIDDEN)
w2 = np.random.randn(HIDDEN, D_OUT)
```

```python
ETA = 1e-6
EPOCHS = 500
```

```python
for t in range(EPOCHS):
    # Forward
    h = x.dot(w1)
    h_relu = np.maximum(h, 0)
    y_pred = h_relu.dot(w2)
    
    # Loss (SSE)
    loss = np.square(y_pred - y).sum()
    print('%3d: %11.2f' % (t, loss))
    
    # Back
    grad_y_pred = 2. * (y_pred - y)
    grad_w2 = h_relu.T.dot(grad_y_pred)
    grad_h_relu = grad_y_pred.dot(w2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h < 0] = 0
    grad_w1 = x.T.dot(grad_h)
    
    # Update weights
    w1 -= ETA * grad_w1
    w2 -= ETA * grad_w2
```
